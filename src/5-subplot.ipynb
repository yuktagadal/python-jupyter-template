### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt

# Data for Line Plot
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 18, 23]

# Data for Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

# Data for Scatter Plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

# Data for Pie Chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create Subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line Plot
axs[0, 0].plot(x, y, marker='o', color='blue')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')

# Bar Plot
axs[0, 1].bar(categories, values, color=['blue', 'green', 'red', 'purple'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Scatter Plot
axs[1, 0].scatter(x_scatter, y_scatter, color='orange')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')

# Pie Chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')

# Adjust Layout
plt.tight_layout()

# Show Plot
plt.show()
```
